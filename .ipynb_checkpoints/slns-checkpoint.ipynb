```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# ======================
# (a) Load datasets
# ======================
# Note: These files are local to your machine. The code will not run without them.
normal = pd.read_csv("/Users/wbandabarragan/Library/CloudStorage/Dropbox/GitHub_Repos/physics-teaching-data/1D-data/sample1_hres.csv")["value"].values
lognormal = pd.read_csv("/Users/wbandabarragan/Library/CloudStorage/Dropbox/GitHub_Repos/physics-teaching-data/1D-data/sample2_hres.csv")["value"].values

# For demonstration, we can create sample data.
#np.random.seed(42)
#normal = np.random.normal(loc=100, scale=10, size=1000)
#lognormal = np.random.lognormal(mean=2, sigma=0.5, size=1000)

datasets = {"Normal candidate": normal, "Lognormal candidate": lognormal}

# ======================
# Helper function
# ======================
def analyze_distribution(data, name):
    print(f"\n--- Analyzing {name} ---")

    # (a) Plot histogram
    plt.figure(figsize=(8,4))
    plt.hist(data, bins=40, density=True, alpha=0.6, label="Data histogram")
    
    # (b) Fit to candidate distributions
    candidates = {
        "norm": stats.norm,      # normal distribution
        "lognorm": stats.lognorm # log-normal distribution
    }
    
    best_fit = None
    best_pval = -1
    
    for dist_name, dist in candidates.items():
        # Fit parameters
        params = dist.fit(data)
        
        # Perform KS test (note: use the SciPy name string, e.g. "norm")
        ks_stat, pval = stats.kstest(data, dist_name, args=params)
        
        # Save best
        if pval > best_pval:
            best_fit = (dist_name, dist, params, pval)
            best_pval = pval

    
    dist_name, dist, params, pval = best_fit
    print(f"Best fit: {dist_name} with p-value={pval:.4f}")
    print(f"Parameters: {params}")
    
    # Plot fitted PDF
    x = np.linspace(min(data), max(data), 500)
    plt.plot(x, dist.pdf(x, *params), 'r-', lw=2, label=f"Best fit ({dist_name})")
    plt.legend()
    plt.title(f"{name} - Best Fit: {dist_name}")
    plt.savefig(f"{name}_best_fit.png")
    
    # (c) Mean and std with CI
    mean = np.mean(data)
    
    # Corrected: Use percentiles to find asymmetric confidence intervals.
    # The percentile values approximate the symmetric confidence levels of a normal distribution.
    conf_levels = {
        "1σ": [16, 84],
        "2σ": [2.5, 97.5],
        "3σ": [0.15, 99.85]
    }
    
    print(f"Mean = {mean:.3f}")
    print("Confidence intervals (from percentiles):")
    
    # Get and print the asymmetric errors
    conf_intervals = {}
    for key, levels in conf_levels.items():
        lo, hi = np.percentile(data, levels)
        conf_intervals[key] = (lo, hi)
        err_low = mean - lo
        err_high = hi - mean
        print(f"  {key}: [{lo:.3f}, {hi:.3f}] or {mean:.3f} +{err_high:.3f}/-{err_low:.3f}")

    # Visualization of CI
    plt.figure(figsize=(8,4))
    plt.hist(data, bins=40, density=True, alpha=0.6, label="Data histogram")
    plt.axvline(mean, color="k", linestyle="--", label="Mean")
    
    # Use the corrected, percentile-based confidence intervals for plotting
    plt.axvspan(*conf_intervals["1σ"], color="yellow", alpha=0.3, label="1σ CI")
    plt.axvspan(*conf_intervals["2σ"], color="orange", alpha=0.2, label="2σ CI")
    plt.axvspan(*conf_intervals["3σ"], color="red", alpha=0.1, label="3σ CI")
    plt.legend()
    plt.title(f"{name} - Confidence Intervals")
    plt.savefig(f"{name}_confidence_intervals.png")


# ======================
# Run analysis
# ======================
for name, data in datasets.items():
    analyze_distribution(data, name)
```
